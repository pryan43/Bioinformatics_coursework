```python
def sub_divide(x):
  print('initial x', x)
  x -= 2
  return x / 2, x / 4

hba1c_column = df['hba1c']
print(hba1c_column)

for v in hba1c_column[::10]:
  vf = sub_divide(v)
  print(vf)

```

```python
#Scatterplot of Dataframe
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.subplots()

ax.scatter(df['rbg'], df['hba1c'], s=100, label="Reported Readings")
ax.set_title('Rbg and Hba1c Readings in Diabetic Patients')
ax.set_xlabel('rbg (mmol/l)')
ax.set_ylabel('hba1c (DCCT percentages)')
ax.legend(bbox_to_anchor=(0, 1), loc='lower right')
plt.show()
```
