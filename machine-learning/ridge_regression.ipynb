#HW 3 Regularizing Linear Regeression

###Importing modules and reading CSV (code cell 1)(From HW 2)

```python
#Code cell 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bloodglucose = pd.read_csv('/content/drive/MyDrive/Machine_Learning/HW_3/Copy of hba1c_bloodGlucose_selected_use_train_test_0.5_seed_31.csv', index_col = 0)
print(bloodglucose.head())
```

###Generating random number and splitting data (code cell 2)(From HW 2)

```python
#Code cell 2
from sklearn.model_selection import train_test_split
np.random.seed(31)
bg_fit, bg_test = train_test_split(bloodglucose, test_size = 0.50)
```

###Fitting to a line (code cell 3) (From HW 2)

```python

#Code cell 3
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


p3 = PolynomialFeatures(3)

x_fit = bg_fit[['rbg_scaled']]
y_fit = bg_fit['hba1c']
x_validation = bg_test[['rbg_scaled']]
y_validation = bg_test['hba1c']


bg_fit_p3 = p3.fit_transform(x_fit)
bg_test_p3 = p3.fit_transform(x_validation)

m = LinearRegression()
m.fit(bg_fit_p3, y_fit)


x_range = np.max(bloodglucose['rbg_scaled']) - np.min(bloodglucose['rbg_scaled'])
model_x = np.linspace(np.min(bloodglucose['rbg_scaled']) - 0.1*x_range, np.max(bloodglucose['rbg_scaled'])+ 0.1*x_range, 100).reshape(-1, 1)
model_x_p3 = p3.fit_transform(model_x)


figsize = (10, 7)
fig, ax = plt.subplots(figsize=figsize)


ax.scatter(x_fit, y_fit, s=100, label='fitting data')
ax.scatter(x_validation, y_validation, s=100, label='validation data')

ax.plot(model_x, m.predict(model_x_p3), label='Fit (degree 3)')



ax.set_xlabel('rbg_scaled')
ax.set_ylabel('hba1c')
ax.legend()


plt.show()
```

###R^2 Score for fitting and testing data (code cell 4)(From HW 2)

```python
#Code cell 4

fit_score = m.score(p3.fit_transform(x_fit),y_fit )
print('Score for fitting data (degree 3):', fit_score)

validation_score = m.score(p3.fit_transform(x_validation),y_validation )
print('Score for validation data (degree 3):', validation_score)

```

###Code cell 4 discussion
#####I think that this model is underfit since it is not really flexible enough to reach all or the majortiy of the data points. I believe that there is room for improvement and adding flexibilty or more parameters could improve the model. The score shows that the model is currently okay by being around 0.61 for both the fit score and validation score. This is a little over halfway between 0(model no better than mean) and 1 (perfect model), so it is not bad but has room for improvement. The scores are relatively the same between the fit data and validation data and this is a good sign since it means the model is behaving similarly for both sets of data.

###Fit to polynomial degree 10 (code cell 5)(From HW 2)

```python
#Code cell 5
p10 = PolynomialFeatures(10)

x_fit = bg_fit[['rbg_scaled']]
y_fit = bg_fit['hba1c']
x_validation = bg_test[['rbg_scaled']]
y_validation = bg_test['hba1c']


bg_fit_p10 = p10.fit_transform(x_fit)
bg_test_p10 = p10.fit_transform(x_validation)

m = LinearRegression()
m.fit(bg_fit_p10, y_fit)


x_range = np.max(bloodglucose['rbg_scaled']) - np.min(bloodglucose['rbg_scaled'])
model_x = np.linspace(np.min(bloodglucose['rbg_scaled']) - 0.1*x_range, np.max(bloodglucose['rbg_scaled'])+ 0.1*x_range, 100).reshape(-1, 1)
model_x_p10 = p10.fit_transform(model_x)


figsize = (10, 7)
fig, ax = plt.subplots(figsize=figsize)


ax.scatter(x_fit, y_fit, s=100, label='fitting data')
ax.scatter(x_validation, y_validation, s=100, label='validation data')

ax.plot(model_x, m.predict(model_x_p10), label='Fit (degree 10)')



ax.set_xlabel('rbg_scaled')
ax.set_ylabel('hba1c')
ax.set_ylim(0, 15)
ax.legend()


plt.show()

fit_score = m.score(p10.fit_transform(x_fit),y_fit )
print('Score for fitting data (degree 10):', fit_score)

validation_score = m.score(p10.fit_transform(x_validation),y_validation)
print('Score for validation data (degree 10):', validation_score)
```

###Code cell 5 discussion

#####The graph and scores shown above are an example of overfitting. You can see how the model is overfit in the graph because it aligns well with the data points but then outside of the data drops above and down in the beginning and end of the data set and has a bit of a strange curve between data points in the beginning. The scores also show us that the model is overfit because the score for the fitting data drastically improved to a 0.93 but at the cost of the validation data score which became -22.39 which is so much worse than 0 (meaning it is an even worse model than just predicting the mean).

###Start of HW 3: Testing Regularization

```python
#Code cell 1:
from sklearn.linear_model import Ridge


p10 = PolynomialFeatures(10)
p10.fit(bg_fit[['rbg_scaled']])

x_fit = bg_fit[['rbg_scaled']]
y_fit = bg_fit['hba1c']
x_validation = bg_test[['rbg_scaled']]
y_validation = bg_test['hba1c']


bg_fit_p10 = p10.transform(x_fit)
bg_test_p10 = p10.transform(x_validation)

m = LinearRegression()
m.fit(bg_fit_p10, y_fit)

ridge_model = Ridge(0.01)
ridge_model.fit(p10.transform(bg_fit[['rbg_scaled']]), bg_fit['hba1c'])


x_range = np.max(bloodglucose['rbg_scaled']) - np.min(bloodglucose['rbg_scaled'])
model_x = np.linspace(np.min(bloodglucose['rbg_scaled']) - 0.1*x_range, np.max(bloodglucose['rbg_scaled'])+ 0.1*x_range, 100).reshape(-1, 1)
model_x_p10 = p10.transform(model_x)


figsize = (10, 7)
fig, ax = plt.subplots(figsize=figsize)
ax.scatter(x_fit, y_fit, s=100, color='grey', label='fitting data')
ax.plot(model_x, m.predict(model_x_p10), lw=4, label='ordinary least sq.')
ax.plot(model_x, ridge_model.predict(model_x_p10), '--', dashes=(2,3), lw=5, label='ridge (no regulariz.)')


ax.set_xlabel('rbg_scaled')
ax.set_ylabel('hba1c')
ax.set_ylim(0,15)
ax.legend()


plt.show()

fit_score = m.score(p10.fit_transform(x_fit),y_fit )
print('Score for fitting data (degree 10):', fit_score)
```


###Alpha Value that is too small (Code cell 2)
  This alpha value is too small and you can tell by the graph as well as the score. The ridge is indicated in the orange dashed line and you can see that with an alpha value of 0.000001 (very close to zero) it still has some pretty big oscillations that are similar to the fit model. This is an example of overfitting since it would provide a really good score for the fitting data, but not a good score for other data sets. You can also see that the score given to the ridge value when compared against the validation data has a score of 0.377 which is sort of close to 0, meaning it is not a very good model.


```python
#Code cell 2:

from sklearn.linear_model import Ridge


p10 = PolynomialFeatures(10)
p10.fit(bg_fit[['rbg_scaled']])

x_fit = bg_fit[['rbg_scaled']]
y_fit = bg_fit['hba1c']
x_validation = bg_test[['rbg_scaled']]
y_validation = bg_test['hba1c']


bg_fit_p10 = p10.transform(x_fit)
bg_test_p10 = p10.transform(x_validation)

m = LinearRegression()
m.fit(bg_fit_p10, y_fit)

ridge_model = Ridge(0.000001)
ridge_model.fit(p10.transform(bg_fit[['rbg_scaled']]), bg_fit['hba1c'])


x_range = np.max(bloodglucose['rbg_scaled']) - np.min(bloodglucose['rbg_scaled'])
model_x = np.linspace(np.min(bloodglucose['rbg_scaled']) - 0.1*x_range, np.max(bloodglucose['rbg_scaled'])+ 0.1*x_range, 100).reshape(-1, 1)
model_x_p10 = p10.transform(model_x)

figsize = (10, 7)
fig, ax = plt.subplots(figsize=figsize)
ax.scatter(x_fit, y_fit, s=100, color='grey', label='fitting data')
ax.plot(model_x, m.predict(model_x_p10), lw=4, label='ordinary least sq.')
ax.plot(model_x, ridge_model.predict(model_x_p10), '--', dashes=(2,3), lw=5, label='ridge (no regulariz.)')



ax.set_xlabel('rbg_scaled')
ax.set_ylabel('hba1c')
ax.set_ylim(0,15)
ax.legend()


plt.show()

fit_score = m.score(p10.transform(x_fit),y_fit )
print('Score for fitting data (degree 10):', fit_score)
ridge_score = ridge_model.score(p10.transform(x_validation),y_validation)
print('Score for ridge (alpha value 0.000001):', ridge_score)
```

###Alpha Value that is too big (code cell 3)
If I change the alpha value to 1000, this is an example of when it is too big. When the alpha gets bigger and bigger the line will continue to get straighter. The line gets flatter and smoother the larger the alpha value gets. This line for the alpha score is relatively straight with a cuve at the end, almost representing an exponential line. This is an example of underfitting because the line is not flexible enough to represent the data points. Similar to the overfitting model, the score given to the ridge value when compared against the validation data has a score of 0.173 which is even closer to 0, making it an even worse model than the overfitting one.

```python
#Code cell 3:

from sklearn.linear_model import Ridge


p10 = PolynomialFeatures(10)
p10.fit(bg_fit[['rbg_scaled']])

x_fit = bg_fit[['rbg_scaled']]
y_fit = bg_fit['hba1c']
x_validation = bg_test[['rbg_scaled']]
y_validation = bg_test['hba1c']


bg_fit_p10 = p10.transform(x_fit)
bg_test_p10 = p10.transform(x_validation)

m = LinearRegression()
m.fit(bg_fit_p10, y_fit)

ridge_model = Ridge(10000)
ridge_model.fit(p10.transform(bg_fit[['rbg_scaled']]), bg_fit['hba1c'])


x_range = np.max(bloodglucose['rbg_scaled']) - np.min(bloodglucose['rbg_scaled'])
model_x = np.linspace(np.min(bloodglucose['rbg_scaled']) - 0.1*x_range, np.max(bloodglucose['rbg_scaled'])+ 0.1*x_range, 100).reshape(-1, 1)
model_x_p10 = p10.transform(model_x)


figsize = (10, 7)
fig, ax = plt.subplots(figsize=figsize)
ax.scatter(x_fit, y_fit, s=100, color='grey', label='fitting data')
ax.plot(model_x, m.predict(model_x_p10), lw=4, label='ordinary least sq.')
ax.plot(model_x, ridge_model.predict(model_x_p10), '--', dashes=(2,3), lw=5, label='ridge (no regulariz.)')



ax.set_xlabel('rbg_scaled')
ax.set_ylabel('hba1c')
ax.set_ylim(0,15)
ax.legend()


plt.show()

fit_score = m.score(p10.transform(x_fit),y_fit)
print('Score for fitting data (degree 10):', fit_score)
ridge_score = ridge_model.score(p10.transform(x_validation),y_validation)
print('Score for ridge (alpha value 10000):', ridge_score)
```

###Alpha Value that is just right (code cell 4)
  This alpha value is just right. Setting the alpha value to 1 seems like the best value for this data set. The graph shows the how the line gently curves between data points and gives a good average representation of the data. The score of the alpha value is 0.567 which is right in between 0  (bad model)and 1 (perfect model) so it is an okay score. The score goes down ever so sligthly when you shift it to 10 or 0.1, which are comparable, but an alpha value of 1 is the best for this model.
